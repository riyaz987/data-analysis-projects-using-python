```python
df_africa=df['country']=='Africa'
africa_ids=df['country'].index[df_africa]
africa_ids=list(africa_ids)
africa_clean = df.loc[africa_ids,:].copy()
sns.lineplot(x='year',y='co2',data=africa_clean).set_title('Africa CO2 Emission') 
```

```python
df_europe=df['country']=='Europe'
europe_ids=df['country'].index[df_europe]
europe_ids=list(europe_ids)
europe_clean = df.loc[europe_ids,:].copy()
sns.lineplot(x='year',y='co2',data=europe_clean).set_title('Europe CO2 Emission') 
```

```python
df_north_america=df['country']=='North America'
north_america_ids=df['country'].index[df_north_america]
north_america_ids=list(north_america_ids)
north_america_clean = df.loc[north_america_ids,:].copy()
sns.lineplot(x='year',y='co2',data=north_america_clean).set_title('North America CO2 Emission') 
```

```python
df_ocean=df['country']=='Oceania'
ocean_ids=df['country'].index[df_ocean]
ocean_ids=list(ocean_ids)
ocean_clean = df.loc[ocean_ids,:].copy()
sns.lineplot(x='year',y='co2',data=ocean_clean).set_title('Ocean CO2 Emission')  
```
